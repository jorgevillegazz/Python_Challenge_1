# 3. Calificación promedio de la tienda

```python
total_calificaciones_1 = tienda['Calificación'].sum()                     #se crea la variable  con el total de las calificaciones
cantidad_de_productos_1= len(tienda['Producto'])                          #se crea una vriable con el total de los productos

promedio_calificacion_1 = total_calificaciones_1/cantidad_de_productos_1  #se saca el promedio de la calificacion de la tienda
promedio_calificacion_1 = round(promedio_calificacion_1, 2)               #redondeamos el resuultado el promedio

print(f"El promedio de la calificación de los clientes en la tienda uno es: {promedio_calificacion_1}")
```
