```python
urls = {
    "tienda_1": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv",
    "tienda_2": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv",
    "tienda_3": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv",
    "tienda_4": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"
}
```

```python
#Cargando Datos
datos_tienda = {}

for nombre_tienda, url in urls.items():
    datos_tienda[nombre_tienda] = pd.read_csv(url)
    print(f"✓ {nombre_tienda} Cargando: {len(datos_tienda[nombre_tienda])} registros")

print("\n" + "="*50)
print("Análisis esploratorio de las Tiendas")
print("="*50)
```

```python
for nombre_tienda, df in datos_tienda.items():
    ventas_por_categoria = df.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
    print(f"\n{nombre_tienda}:")
    for categoria, cantidad in ventas_por_categoria.items():
        print(f"  {categoria}: {cantidad} productos")
```

```python
# Gráfico consolidado por categoria
plt.figure(figsize=(14, 8))

# Concatenar todos os dados
todas_las_tiendas_df = pd.concat(datos_tienda.values(), keys=datos_tienda.keys(), names=['Tienda', None])

# Agrupar por loja e categoria
ventas_por_categoria_por_Tienda = todas_las_tiendas_df.groupby(['Tienda', 'Categoría del Producto'])['Producto'].count().unstack(fill_value=0)

# Plotar
ax = ventas_por_categoria_por_Tienda.plot(kind='bar', width=0.8, color=['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEAA7'])
ax.set_title('Ventas por Categoria en cada Tienda', fontsize=16, fontweight='bold')
ax.set_xlabel('Tiendas')
ax.set_ylabel('Unidades Vendidas')
ax.tick_params(axis='x', rotation=0)
ax.legend(title='Categoría del Producto', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
