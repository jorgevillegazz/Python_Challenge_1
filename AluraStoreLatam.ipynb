```python
# Productos m√°s y menos vendidos

diccionario_tiendas = {
    "Tienda 1": tienda,
    "Tienda 2": tienda2,
    "Tienda 3": tienda3,
    "Tienda 4": tienda4
}

for nombre, df in diccionario_tiendas.items():
    print(f"\n{nombre} - Productos m√°s y menos vendidos:")

    productos_vendidos = df["Producto"].value_counts()

    producto_mas = productos_vendidos.idxmax()
    cantidad_mas = productos_vendidos.max()

    producto_menos = productos_vendidos.idxmin()
    cantidad_menos = productos_vendidos.min()

    print(f"  M√°s vendido: {producto_mas} ({cantidad_mas} ventas)")
    print(f"  Menos vendido: {producto_menos} ({cantidad_menos} ventas)")
```

# 5. Env√≠o promedio por tienda

En este paso, debes calcular el costo de env√≠o promedio para cada tienda. El objetivo es comprender cu√°nto se gasta, en promedio, en el env√≠o de cada tienda.  Sabiendo que el cliente paga el env√≠o.

```python
total_costo_envio_Tienda1 = tienda['Costo de env√≠o'].sum()                     #se crea la variable  con el total de las calificaciones
cantidad_de_productos_Tienda1= len(tienda['Producto'])                          #se crea una vriable con el total de los productos

promedio_costo_envio_Tienda1 = total_costo_envio_Tienda1/cantidad_de_productos_Tienda1  #se saca el promedio de la calificacion de la tienda
promedio_costo_envio_Tienda1 = round(promedio_costo_envio_Tienda1, 2)               #redondeamos el resuultado el promedio

print(f"El promedio de la calificaci√≥n de los clientes en la tienda uno es: {promedio_costo_envio_Tienda1}")
```

```python
# optimizando el codigo para su resultado

diccionario_tiendas = {
    "Tienda 1": tienda,
    "Tienda 2": tienda2,
    "Tienda 3": tienda3,
    "Tienda 4": tienda4
}

nombres_tiendas = []
promedios_envio = []

for nombre, df in diccionario_tiendas.items():
    promedio_envio = df["Costo de env√≠o"].mean()
    nombres_tiendas.append(nombre)
    promedios_envio.append(promedio_envio)
    print(f"{nombre} - Costo de env√≠o promedio: ${promedio_envio:,.2f}")
```
